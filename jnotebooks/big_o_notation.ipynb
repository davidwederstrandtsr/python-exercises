## Curves to graph
$y = 0x + 1$ and label the curve "O(1)"


$y = log(x)$ and label the curve "O(log n)"


$y = x$ and label the curve "O(n)"

$y = x * log(x)$ and label it "O(n log n)"

$y = x^2$ and label it "O(n^2)"

$y = 2^x$ and label it "O(2^n)"

$y = n!$ and label is "O(n!)"

$y = n^n$ and label it "O(n^n)"

```python
get_ipython().run_line_magic('matplotlib', 'inline')

import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
x = range(1, 50)

y1 = [0 * n + 1 for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x]
y5 = [n**2 for n in x]
y6 = [2**n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n**n for n in x]
```

```python
plt.figure(figsize=(16, 10))
plt.plot(x, y1,c="blue", label='$y = 0x + 1$')
plt.plot(x, y2, c="red", label='$y = log(x)$')
plt.plot(x, y3, c="green", label='$y = x$')
plt.plot(x, y4, c="orange", label='$y = x * log(x)$')
plt.plot(x, y5,c="yellow", label='$y = x^2$')
plt.plot(x, y6, c="black", label='$y = 2^x$')
plt.plot(x, y7, c="cyan", label='$y = n!$')
plt.plot(x, y8, c="purple", label='$y = n^n$')
plt.xlabel('Elements', fontsize=12)
plt.ylabel('Operations', fontsize=12)
plt.title('Big O Notation', fontsize=16)
plt.ylim(0, 15)
plt.xlim(1, 10)
plt.legend()
plt.show()
```
